```python
import pynattas as pnas
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import transforms
from pycocotools.coco import COCO
import os
import rasterio
from PIL import Image, ImageDraw
import math
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import median_filter, gaussian_filter
```

```python
def inference(original_img, preds, obj_threshold=0.5, iou_threshold=0.3, min_area=3, max_area=4000, min_AR=1/10, max_AR=10, show_image=True):
    # Remove boxes with area less than 2 pixels^2 in (cx, cy, w, h) format
    device=preds.device
    preds = remove_small_boxes(preds, min_area=min_area/(512*512))
    preds = remove_big_boxes(preds, max_area=max_area/(512*512))
    preds = remove_big_aspect_ratio(preds, min_AR=1/10, max_AR=10)

    if preds.numel() == 0:
        return preds

    # Filter out low objectness predictions
    #objectness_scores = torch.sigmoid(preds[..., 4])
    objectness_scores = preds[..., 4]
    keep = objectness_scores > obj_threshold
    preds = preds[keep]

    if preds.numel() == 0:
        return preds

    # Convert from cx, cy, w, h to x1, y1, x2, y2 before further processing
    preds = cxcywh_to_xyxy(preds)

    # Apply Non-Max Suppression (NMS)
    preds = apply_nms(preds, iou_threshold)

    if preds.numel() == 0:
        return preds

    img_size = original_img.shape  # Width, Height
    preds[..., :4] = scale_boxes(preds[..., :4].cpu(), img_size)

    return preds

def cxcywh_to_xyxy(preds):
    """
    Convert bounding boxes from (cx, cy, w, h) format to (x1, y1, x2, y2) format.
    
    Args:
        preds: Tensor of shape (N, 6) where N is the number of boxes, 
               each box has format [cx, cy, w, h, score, class].
    
    Returns:
        Tensor of same shape with (x1, y1, x2, y2, score, class).
    """
    preds = preds.clone()  # Clone to avoid modifying the original tensor
    
    # Convert from cx, cy, w, h to x1, y1, x2, y2
    preds[:, 0] = preds[:, 0] - preds[:, 2] / 2  # x1 = cx - w/2
    preds[:, 1] = preds[:, 1] - preds[:, 3] / 2  # y1 = cy - h/2
    preds[:, 2] = preds[:, 0] + preds[:, 2]      # x2 = cx + w/2
    preds[:, 3] = preds[:, 1] + preds[:, 3]      # y2 = cy + h/2
    
    return preds


def apply_nms(preds, iou_threshold):
    bboxes = preds[..., :4]
    scores = preds[..., 4]
    classes = preds[..., 5:].argmax(dim=-1)

    keep_boxes = []
    unique_classes = classes.unique()
    for cls in unique_classes:
        cls_mask = classes == cls
        cls_bboxes = bboxes[cls_mask]
        cls_scores = scores[cls_mask]

        keep = nms(cls_bboxes, cls_scores, iou_threshold)
        keep_boxes.append(preds[keep])

    if len(keep_boxes) == 0:
        return torch.stack(keep_boxes, dim=0)
    
    return torch.cat(keep_boxes, dim=0)

def nms(bboxes, scores, iou_threshold):
    idxs = scores.argsort(descending=True)
    keep = []

    while idxs.numel() > 0:
        i = idxs[0]
        keep.append(i)

        if idxs.numel() == 1:
            break

        ious = torch.tensor([iou_single(bboxes[i], bboxes[j]) for j in idxs[1:]], device=bboxes.device)
        #ious = bboxes_iou(bboxes[i].unsqueeze(0), bboxes[idxs[1:]])
        idxs = idxs[1:][ious <= iou_threshold]

    return torch.tensor(keep, dtype=torch.long, device=bboxes.device)

def iou_single(box1, box2):
    inter_rect_x1 = max(box1[0], box2[0])
    inter_rect_y1 = max(box1[1], box2[1])
    inter_rect_x2 = min(box1[2], box2[2])
    inter_rect_y2 = min(box1[3], box2[3])

    inter_area = max(inter_rect_x2 - inter_rect_x1 + 1/512, 0) * max(inter_rect_y2 - inter_rect_y1 + 1/512, 0)
    b1_area = (box1[2] - box1[0] + 1/512) * (box1[3] - box1[1] + 1/512)
    b2_area = (box2[2] - box2[0] + 1/512) * (box2[3] - box2[1] + 1/512)

    iou = inter_area / (b1_area + b2_area - inter_area + 1e-9)
    return iou

def bboxes_iou(box1, box2):
    inter_rect_x1 = torch.max(box1[:, 0], box2[:, 0])
    inter_rect_y1 = torch.max(box1[:, 1], box2[:, 1])
    inter_rect_x2 = torch.min(box1[:, 2], box2[:, 2])
    inter_rect_y2 = torch.min(box1[:, 3], box2[:, 3])

    inter_area = torch.clamp(inter_rect_x2 - inter_rect_x1 + 1/512, min=0) * torch.clamp(inter_rect_y2 - inter_rect_y1 + 1/512, min=0)
    b1_area = (box1[:, 2] - box1[:, 0] + (1/512)) * (box1[:, 3] - box1[:, 1] + (1/512))
    b2_area = (box2[:, 2] - box2[:, 0] + (1/512)) * (box2[:, 3] - box2[:, 1] + (1/512))

    iou = inter_area / (b1_area + b2_area - inter_area + 1e-9)
    return iou

def scale_boxes(boxes, img_size):
    width, height = img_size

    for i in range(len(boxes)):
        boxes[i, 0] = boxes[i, 0] * width #int(math.floor(boxes[i, 0] * width))
        boxes[i, 1] = boxes[i, 1] * height #int(math.floor(boxes[i, 1] * height))
        boxes[i, 2] = boxes[i, 2] * width #int(math.ceil(boxes[i, 2] * width))
        boxes[i, 3] = boxes[i, 3] * height #int(math.ceil(boxes[i, 3] * height))
    
    return boxes

'''
def scale_boxes(boxes, img_size):
    if boxes.numel() == 0:
        return boxes  # Return empty tensor if boxes are empty

    width, height = img_size[1], img_size[0]  # img_size should be (height, width)

    # Ensure boxes is a float tensor for multiplication
    boxes = boxes.float()

    # Vectorized scaling
    boxes[:, 0] *= width
    boxes[:, 1] *= height
    boxes[:, 2] *= width
    boxes[:, 3] *= height

    return boxes
'''

def do_boxes_intersect(box1,box2,patience=10):
     inter_rect_x1 = max(box1[0], box2[0])-patience
     inter_rect_y1 = max(box1[1], box2[1])-patience
     inter_rect_x2 = min(box1[2], box2[2])+patience
     inter_rect_y2 = min(box1[3], box2[3])+patience

     inter_area = max((inter_rect_x2) - (inter_rect_x1+10) + 1, 0) * max(inter_rect_y2 - inter_rect_y1 + 1, 0)

     if inter_area > 0:
         return True
     else:    
         return False

'''def merge_intersecting_boxes(preds,patience=10): # Does not do classwise merging, assumes all boxes are of the same class, for now
    preds = preds.clone()
    preds_list = preds.tolist()
    print(f"Number of boxes before matching: {len(preds_list)}")

    i = 0
    while i < (len(preds_list)-1):
        #print(f"Checking box i:{i}")
        j = i+1
        while j < (len(preds_list)):
            #print(f"Checking box j:{j}")
            if do_boxes_intersect(preds_list[i][:4], preds_list[j][:4], patience):
                x1 = min(preds_list[i][0], preds_list[j][0])
                y1 = min(preds_list[i][1], preds_list[j][1])
                x2 = max(preds_list[i][2], preds_list[j][2])
                y2 = max(preds_list[i][3], preds_list[j][3])
                score = (preds_list[i][4] + preds_list[j][4]) / 2
                preds_list[i] = [x1, y1, x2, y2, score, preds_list[i][5:]]
                preds_list.pop(j)
                j = i
            j += 1
                
        i += 1

    print(f"Number of boxes after matching: {len(preds_list)}")
        
    return preds_list'''

def merge_intersecting_boxes(preds, patience=10):
    """
    Merges intersecting bounding boxes using tensors for efficiency.
    Assumes all boxes are of the same class for now.

    Args:
        preds (Tensor): Tensor of shape (N, 5 + num_classes), where N is the number of boxes.
                        Each box has the format [x1, y1, x2, y2, score, class_scores...].
        patience (int): Margin added to the boxes to determine intersection.

    Returns:
        Tensor: Tensor of merged boxes.
    """
    if preds.numel() == 0:
        return preds

    device = preds.device
    boxes = preds[:, :4]
    scores = preds[:, 4]
    class_scores = preds[:, 5:]

    # Adjust boxes with patience
    boxes_adj = torch.zeros_like(boxes)
    boxes_adj[:, 0] = boxes[:, 0] - patience  # x1 - patience
    boxes_adj[:, 1] = boxes[:, 1] - patience  # y1 - patience
    boxes_adj[:, 2] = boxes[:, 2] + patience  # x2 + patience
    boxes_adj[:, 3] = boxes[:, 3] + patience  # y2 + patience

    N = boxes.size(0)
    merged = torch.zeros(N, dtype=torch.bool, device=device)
    clusters = []

    for i in range(N):
        if merged[i]:
            continue
        # Initialize cluster with box i
        cluster_indices = [i]
        merged[i] = True
        # Compare box i with all other boxes
        for j in range(i + 1, N):
            if merged[j]:
                continue
            if boxes_intersect(boxes_adj[i], boxes_adj[j]):
                cluster_indices.append(j)
                merged[j] = True
        # Merge boxes in the cluster
        cluster_boxes = preds[cluster_indices]
        x1 = cluster_boxes[:, 0].min()
        y1 = cluster_boxes[:, 1].min()
        x2 = cluster_boxes[:, 2].max()
        y2 = cluster_boxes[:, 3].max()
        score = cluster_boxes[:, 4].mean()
        class_score = cluster_boxes[:, 5:].mean(dim=0)
        merged_box = torch.cat([x1.unsqueeze(0), y1.unsqueeze(0), x2.unsqueeze(0), y2.unsqueeze(0),
                                score.unsqueeze(0), class_score])
        clusters.append(merged_box)

    if len(clusters) == 0:
        return torch.empty((0, preds.size(1)), device=device)

    preds_merged = torch.stack(clusters, dim=0)
    return preds_merged

def boxes_intersect(box1, box2):
    """
    Checks if two boxes intersect.

    Args:
        box1 (Tensor): Tensor of shape (4,), [x1, y1, x2, y2].
        box2 (Tensor): Tensor of shape (4,), [x1, y1, x2, y2].

    Returns:
        bool: True if boxes intersect, False otherwise.
    """
    inter_x1 = torch.max(box1[0], box2[0])
    inter_y1 = torch.max(box1[1], box2[1])
    inter_x2 = torch.min(box1[2], box2[2])
    inter_y2 = torch.min(box1[3], box2[3])

    inter_w = torch.clamp(inter_x2 - inter_x1 + 1, min=0)
    inter_h = torch.clamp(inter_y2 - inter_y1 + 1, min=0)
    inter_area = inter_w * inter_h

    return inter_area > 0


def draw_boxes(image, preds, labels):
    # Define category names
    category_names = {
        0: "ship",
    }
    
    image = (image - image.min()) / (image.max() - image.min())
    image = image ** 0.3
    
    plt.figure(figsize=(15, 5))
    plt.subplot(1,3,1); plt.title("Original Image")
    plt.axis('off')
    plt.imshow(image, cmap='gray')

    # Convert to a PIL image for drawing
    image = Image.fromarray((image * 255).astype(np.uint8))
    image = image.convert('RGB')
    draw = ImageDraw.Draw(image)
    #font = ImageFont.truetype('arial.ttf', 15)

    for idx, label in enumerate(labels):
        x1, y1, x2, y2 = label
        draw.rectangle([x1, y1, x2, y2], outline="yellowgreen", width=4)
        print(f"Label {idx}: {label}")
        #draw.text((x1, y1), label, fill="black")

    plt.subplot(1,3,2); plt.title("Ground Truth")
    plt.axis('off')
    plt.imshow(image)

    for pred in preds:
        x1, y1, x2, y2 = pred[:4]
        #x1 = xc - w/2
        #y1 = yc - w/2
        #x2 = xc + w/2
        #y2 = yc + w/2
        #class_id = pred[5:].argmax().item()
        #label = category_names[class_id]
        draw.rectangle([x1, y1, x2, y2], outline="crimson", width=4)
        #draw.text((x1, y1), str(pred[4]), fill="crimson")

    plt.subplot(1,3,3); plt.title("Predictions")
    plt.imshow(image)
    plt.axis('off')
    plt.show()

def remove_small_boxes(preds, min_area=2):
    """
    Removes bounding boxes that have an area smaller than the given threshold in (cx, cy, w, h) format.
    
    Args:
        preds: Tensor of shape (N, 6) where N is the number of boxes, 
               each box has format [cx, cy, w, h, score, class].
        min_area: Minimum area in pixels^2 for the boxes to be kept.
        
    Returns:
        Filtered predictions where each box has area >= min_area.
    """
    # w and h are the width and height in the (cx, cy, w, h) format
    w = preds[..., 2]
    h = preds[..., 3]
    
    # Calculate area of each bounding box
    areas = w * h
    
    # Filter out boxes with area < min_area
    keep = areas >= min_area
    return preds[keep]

def remove_big_boxes(preds, max_area=4000):
    """
    Removes bounding boxes that have an area smaller than the given threshold in (cx, cy, w, h) format.
    
    Args:
        preds: Tensor of shape (N, 6) where N is the number of boxes, 
               each box has format [cx, cy, w, h, score, class].
        min_area: Minimum area in pixels^2 for the boxes to be kept.
        
    Returns:
        Filtered predictions where each box has area >= min_area.
    """
    # w and h are the width and height in the (cx, cy, w, h) format
    w = preds[..., 2]
    h = preds[..., 3]
    
    # Calculate area of each bounding box
    areas = w * h
    
    # Filter out boxes with area < min_area
    keep = areas <= max_area
    return preds[keep]

def remove_big_aspect_ratio(preds, min_AR=1/10, max_AR=10):
    """
    Removes bounding boxes that have an area smaller than the given threshold in (cx, cy, w, h) format.
    
    Args:
        preds: Tensor of shape (N, 6) where N is the number of boxes, 
               each box has format [cx, cy, w, h, score, class].
        min_area: Minimum area in pixels^2 for the boxes to be kept.
        
    Returns:
        Filtered predictions where each box has area >= min_area.
    """
    # w and h are the width and height in the (cx, cy, w, h) format
    w = preds[..., 2]
    h = preds[..., 3]
    
    # Calculate area of each bounding box
    ARs = w / h
    
    # Filter out boxes with area < min_area
    keep_max = ARs <= max_AR
    keep_min = ARs >= min_AR
    keep = keep_max & keep_min
    return preds[keep]
```

```python
import json
import os
import torch
from pycocotools.coco import COCO
from collections import defaultdict
import numpy as np
import math
```

```python
# Paths to your test annotations and prediction results
test_ann_file = rf'./data/SPECTRE_prova/split_dataset/annotations/test.json'
predictions_file = rf'./data/SPECTRE_prova/split_dataset/inference_on_test_{code}.json'
```

```python
# Function to compute IoU between two sets of boxes
def compute_iou_matrix(boxes1, boxes2):
    """
    Computes the IoU between each pair of boxes from boxes1 and boxes2.
    Args:
        boxes1: Tensor of shape (N, 4) in [x, y, width, height] format.
        boxes2: Tensor of shape (M, 4) in [x, y, width, height] format.
    Returns:
        iou_matrix: Tensor of shape (N, M) containing IoUs.
    """
    N = boxes1.shape[0]
    M = boxes2.shape[0]
    iou_matrix = torch.zeros((N, M))
    
    for i in range(N):
        box1 = boxes1[i]
        x1, y1, w1, h1 = box1
        x1_max = x1 + w1
        y1_max = y1 + h1
        
        for j in range(M):
            box2 = boxes2[j]
            x2, y2, w2, h2 = box2
            x2_max = x2 + w2
            y2_max = y2 + h2
            
            # Compute intersection
            xi1 = max(x1, x2)
            yi1 = max(y1, y2)
            xi2 = min(x1_max, x2_max)
            yi2 = min(y1_max, y2_max)
            inter_area = max(xi2 - xi1, 0) * max(yi2 - yi1, 0)
            
            # Compute union
            box1_area = w1 * h1
            box2_area = w2 * h2
            union_area = box1_area + box2_area - inter_area + 1e-6  # Add epsilon to avoid division by zero
            
            # Compute IoU
            iou = inter_area / union_area
            iou_matrix[i, j] = iou
    
    return iou_matrix

# Function to compute SIoU between two sets of boxes
def calculate_siou_matrix(pred_boxes, gt_boxes, gamma=0.5, kappa=8, eps=1e-6):
    """
    Calculate the Scaled Intersection over Union (SIoU) between sets of bounding boxes.
    Parameters:
    -----------
    pred_boxes: Tensor of shape [N, 4]
        Predicted bounding boxes in (x, y, w, h) format, where (x, y) is the top-left corner.
    gt_boxes: Tensor of shape [M, 4]
        Ground truth bounding boxes in (x, y, w, h) format, where (x, y) is the top-left corner.
    Returns:
    --------
    siou_matrix: Tensor of shape (N, M) containing SIoUs.
    """
    N = pred_boxes.shape[0]
    M = gt_boxes.shape[0]
    siou_matrix = torch.zeros((N, M))
    
    for i in range(N):
        pred_box = pred_boxes[i].unsqueeze(0)  # Shape (1, 4)
        x1, y1, w1, h1 = pred_box[:, 0], pred_box[:, 1], pred_box[:, 2], pred_box[:, 3]
        x1_max, y1_max = x1 + w1, y1 + h1
        
        for j in range(M):
            gt_box = gt_boxes[j].unsqueeze(0)  # Shape (1, 4)
            x2, y2, w2, h2 = gt_box[:, 0], gt_box[:, 1], gt_box[:, 2], gt_box[:, 3]
            x2_max, y2_max = x2 + w2, y2 + h2
            
            # Compute intersection
            inter_xmin = torch.max(x1, x2)
            inter_ymin = torch.max(y1, y2)
            inter_xmax = torch.min(x1_max, x2_max)
            inter_ymax = torch.min(y1_max, y2_max)
            inter_width = torch.clamp(inter_xmax - inter_xmin, min=0)
            inter_height = torch.clamp(inter_ymax - inter_ymin, min=0)
            inter_area = inter_width * inter_height

            area1 = w1 * h1
            area2 = w2 * h2
            union_area = area1 + area2 - inter_area

            # IoU calculation
            iou = inter_area / (union_area + eps)
            iou = torch.nan_to_num(iou, nan=0.0, posinf=1.0, neginf=0.0)

            # Size factor calculation
            size_factor = torch.sqrt(area1 + area2)
            size_factor = torch.nan_to_num(size_factor, nan=0.0, posinf=1.0, neginf=0.0)

            p = 1 - gamma * torch.exp(- size_factor / (math.sqrt(2)*kappa))

            # SIoU calculation
            siou = torch.pow(iou + eps, p)  # Add eps to IoU to prevent pow(0, p) when IoU is zero

            siou_matrix[i, j] = siou.item()
    
    return siou_matrix
```

```python
# IoU threshold for considering a detection as true positive# Thresholds for considering a detection as true positive
iou_threshold = 0.5  # IoU threshold
siou_threshold = iou_threshold  # SIoU threshold

# Load the test annotations
coco_gt = COCO(test_ann_file)

# Load the predictions
with open(predictions_file, 'r') as f:
    coco_preds = json.load(f)

# Create a mapping from image_id to list of predictions
preds_by_image = defaultdict(list)
for pred in coco_preds:
    preds_by_image[pred['image_id']].append(pred)

# Initialize counters for IoU-based metrics
tp_IoU = 0  # True Positives
fp_IoU = 0  # False Positives
fn_IoU = 0  # False Negatives
ious = []  # List to store IoUs of matched boxes

# Initialize counters for SIoU-based metrics
tp_SIoU = 0  # True Positives
fp_SIoU = 0  # False Positives
fn_SIoU = 0  # False Negatives
siou_list = []  # List to store SIoUs of matched boxes

# Loop over each image in the test set
for img_id in coco_gt.getImgIds():
    # Get ground truth annotations for the image
    ann_ids = coco_gt.getAnnIds(imgIds=img_id)
    anns = coco_gt.loadAnns(ann_ids)
    
    # Extract ground truth boxes
    gt_boxes = []
    for ann in anns:
        bbox = ann['bbox']  # COCO format [x, y, width, height]
        gt_boxes.append(bbox)
    
    gt_boxes = torch.tensor(gt_boxes)
    num_gt_boxes = len(gt_boxes)
    
    # Get predicted boxes for the image
    preds = preds_by_image.get(img_id, [])
    pred_boxes = []
    pred_scores = []
    for pred in preds:
        bbox = pred['bbox']  # COCO format [x, y, width, height]
        score = pred['score']
        pred_boxes.append(bbox)
        pred_scores.append(score)
    
    pred_boxes = torch.tensor(pred_boxes)
    pred_scores = torch.tensor(pred_scores)
    num_pred_boxes = len(pred_boxes)
    
    # Handling images with no ground truth or predictions
    if num_gt_boxes == 0 and num_pred_boxes == 0:
        continue  # No ground truth or predictions; skip image
    
    # Initialize matched indices for IoU and SIoU
    matched_pred_indices_IoU = []
    matched_gt_indices_IoU = []

    matched_pred_indices_SIoU = []
    matched_gt_indices_SIoU = []
    
    # For IoU-based metrics
    if num_gt_boxes == 0:
        # All predictions are false positives
        fp_IoU += num_pred_boxes
    elif num_pred_boxes == 0:
        # All ground truth boxes are missed detections
        fn_IoU += num_gt_boxes
    else:
        # Compute IoU between each predicted box and each ground truth box
        iou_matrix = compute_iou_matrix(pred_boxes, gt_boxes)
        
        # For each predicted box, find the best matching ground truth box
        for i in range(num_pred_boxes):
            # Find the ground truth box with the highest IoU for this prediction
            ious_per_pred = iou_matrix[i]
            max_iou, max_iou_idx = torch.max(ious_per_pred, dim=0)
            
            if max_iou >= iou_threshold:
                if max_iou_idx.item() not in matched_gt_indices_IoU:
                    # This ground truth box hasn't been matched yet
                    tp_IoU += 1
                    matched_pred_indices_IoU.append(i)
                    matched_gt_indices_IoU.append(max_iou_idx.item())
                    ious.append(max_iou.item())
                else:
                    # Ground truth box already matched; count as false positive
                    fp_IoU += 1
            else:
                # IoU is below threshold; count as false positive
                fp_IoU += 1
        
        # Any ground truth boxes not matched are false negatives
        fn_IoU += num_gt_boxes - len(matched_gt_indices_IoU)
    
    # For SIoU-based metrics
    if num_gt_boxes == 0:
        # All predictions are false positives
        fp_SIoU += num_pred_boxes
    elif num_pred_boxes == 0:
        # All ground truth boxes are missed detections
        fn_SIoU += num_gt_boxes
    else:
        # Compute SIoU between each predicted box and each ground truth box
        siou_matrix = calculate_siou_matrix(pred_boxes, gt_boxes)
        
        # For each predicted box, find the best matching ground truth box
        for i in range(num_pred_boxes):
            # Find the ground truth box with the highest SIoU for this prediction
            sious_per_pred = siou_matrix[i]
            max_siou, max_siou_idx = torch.max(sious_per_pred, dim=0)
            
            if max_siou >= siou_threshold:
                if max_siou_idx.item() not in matched_gt_indices_SIoU:
                    # This ground truth box hasn't been matched yet
                    tp_SIoU += 1
                    matched_pred_indices_SIoU.append(i)
                    matched_gt_indices_SIoU.append(max_siou_idx.item())
                    siou_list.append(max_siou.item())
                else:
                    # Ground truth box already matched; count as false positive
                    fp_SIoU += 1
            else:
                # SIoU is below threshold; count as false positive
                fp_SIoU += 1
        
        # Any ground truth boxes not matched are false negatives
        fn_SIoU += num_gt_boxes - len(matched_gt_indices_SIoU)

# Calculate Mean IoU
if len(ious) > 0:
    mean_iou = np.mean(ious)
else:
    mean_iou = 0.0

# Calculate Mean SIoU
if len(siou_list) > 0:
    mean_siou = np.mean(siou_list)
else:
    mean_siou = 0.0

# Print the results
print(f"True Positives (TP_IoU): {tp_IoU}")
print(f"False Positives (FP_IoU): {fp_IoU}")
print(f"False Negatives (FN_IoU): {fn_IoU}")
print(f"Mean IoU: {mean_iou:.4f}")

print(f"True Positives (TP_SIoU): {tp_SIoU}")
print(f"False Positives (FP_SIoU): {fp_SIoU}")
print(f"False Negatives (FN_SIoU): {fn_SIoU}")
print(f"Mean SIoU: {mean_siou:.4f}")

# Calculate Precision, Recall, and F1 Score for IoU
precision_IoU = tp_IoU / (tp_IoU + fp_IoU) if (tp_IoU + fp_IoU) > 0 else 0.0
recall_IoU = tp_IoU / (tp_IoU + fn_IoU) if (tp_IoU + fn_IoU) > 0 else 0.0
f1_score_IoU = 2 * (precision_IoU * recall_IoU) / (precision_IoU + recall_IoU + 1e-6)

print(f"Precision_IoU: {precision_IoU:.4f}")
print(f"Recall_IoU: {recall_IoU:.4f}")
print(f"F1 Score_IoU: {f1_score_IoU:.4f}")

# Calculate Precision, Recall, and F1 Score for SIoU
precision_SIoU = tp_SIoU / (tp_SIoU + fp_SIoU) if (tp_SIoU + fp_SIoU) > 0 else 0.0
recall_SIoU = tp_SIoU / (tp_SIoU + fn_SIoU) if (tp_SIoU + fn_SIoU) > 0 else 0.0
f1_score_SIoU = 2 * (precision_SIoU * recall_SIoU) / (precision_SIoU + recall_SIoU + 1e-6)

print(f"Precision_SIoU: {precision_SIoU:.4f}")
print(f"Recall_SIoU: {recall_SIoU:.4f}")
print(f"F1 Score_SIoU: {f1_score_SIoU:.4f}")
```

```python
import matplotlib.pyplot as plt

min_threshold = 0.0

# Define the range of thresholds to evaluate
thresholds = np.arange(min_threshold, 1.05, 0.05)

# Initialize lists to store metrics for each threshold
precision_IoU_list = []
recall_IoU_list = []
f1_score_IoU_list = []
mean_iou_list = []

precision_SIoU_list = []
recall_SIoU_list = []
f1_score_SIoU_list = []
mean_siou_list = []

# Loop over thresholds
for thresh in thresholds:
    # Initialize counters for IoU-based metrics
    tp_IoU = 0
    fp_IoU = 0
    fn_IoU = 0
    ious = []

    # Initialize counters for SIoU-based metrics
    tp_SIoU = 0
    fp_SIoU = 0
    fn_SIoU = 0
    siou_list = []

    # Loop over each image in the test set
    for img_id in coco_gt.getImgIds():
        # Get ground truth annotations for the image
        ann_ids = coco_gt.getAnnIds(imgIds=img_id)
        anns = coco_gt.loadAnns(ann_ids)
        
        # Extract ground truth boxes
        gt_boxes = []
        for ann in anns:
            bbox = ann['bbox']  # COCO format [x, y, width, height]
            gt_boxes.append(bbox)
        
        gt_boxes = torch.tensor(gt_boxes)
        num_gt_boxes = len(gt_boxes)
        
        # Get predicted boxes for the image
        preds = preds_by_image.get(img_id, [])
        pred_boxes = []
        pred_scores = []
        for pred in preds:
            bbox = pred['bbox']  # COCO format [x, y, width, height]
            score = pred['score']
            pred_boxes.append(bbox)
            pred_scores.append(score)
        
        pred_boxes = torch.tensor(pred_boxes)
        pred_scores = torch.tensor(pred_scores)
        num_pred_boxes = len(pred_boxes)
        
        # Handling images with no ground truth or predictions
        if num_gt_boxes == 0 and num_pred_boxes == 0:
            continue  # No ground truth or predictions; skip image
        
        # Initialize matched indices for IoU and SIoU
        matched_pred_indices_IoU = []
        matched_gt_indices_IoU = []
    
        matched_pred_indices_SIoU = []
        matched_gt_indices_SIoU = []
        
        # For IoU-based metrics
        if num_gt_boxes == 0:
            # All predictions are false positives
            fp_IoU += num_pred_boxes
        elif num_pred_boxes == 0:
            # All ground truth boxes are missed detections
            fn_IoU += num_gt_boxes
        else:
            # Compute IoU between each predicted box and each ground truth box
            iou_matrix = compute_iou_matrix(pred_boxes, gt_boxes)
            
            # For each predicted box, find the best matching ground truth box
            for i in range(num_pred_boxes):
                # Find the ground truth box with the highest IoU for this prediction
                ious_per_pred = iou_matrix[i]
                max_iou, max_iou_idx = torch.max(ious_per_pred, dim=0)
                
                if max_iou >= thresh:
                    if max_iou_idx.item() not in matched_gt_indices_IoU:
                        # This ground truth box hasn't been matched yet
                        tp_IoU += 1
                        matched_pred_indices_IoU.append(i)
                        matched_gt_indices_IoU.append(max_iou_idx.item())
                        ious.append(max_iou.item())
                    else:
                        # Ground truth box already matched; count as false positive
                        fp_IoU += 1
                else:
                    # IoU is below threshold; count as false positive
                    fp_IoU += 1
            
            # Any ground truth boxes not matched are false negatives
            fn_IoU += num_gt_boxes - len(matched_gt_indices_IoU)
        
        # For SIoU-based metrics
        if num_gt_boxes == 0:
            # All predictions are false positives
            fp_SIoU += num_pred_boxes
        elif num_pred_boxes == 0:
            # All ground truth boxes are missed detections
            fn_SIoU += num_gt_boxes
        else:
            # Compute SIoU between each predicted box and each ground truth box
            siou_matrix = calculate_siou_matrix(pred_boxes, gt_boxes)
            
            # For each predicted box, find the best matching ground truth box
            for i in range(num_pred_boxes):
                # Find the ground truth box with the highest SIoU for this prediction
                sious_per_pred = siou_matrix[i]
                max_siou, max_siou_idx = torch.max(sious_per_pred, dim=0)
                
                if max_siou >= thresh:
                    if max_siou_idx.item() not in matched_gt_indices_SIoU:
                        # This ground truth box hasn't been matched yet
                        tp_SIoU += 1
                        matched_pred_indices_SIoU.append(i)
                        matched_gt_indices_SIoU.append(max_siou_idx.item())
                        siou_list.append(max_siou.item())
                    else:
                        # Ground truth box already matched; count as false positive
                        fp_SIoU += 1
                else:
                    # SIoU is below threshold; count as false positive
                    fp_SIoU += 1
            
            # Any ground truth boxes not matched are false negatives
            fn_SIoU += num_gt_boxes - len(matched_gt_indices_SIoU)
    
    # Calculate Precision, Recall, and F1 Score for IoU
    precision_IoU = tp_IoU / (tp_IoU + fp_IoU) if (tp_IoU + fp_IoU) > 0 else 0.0
    recall_IoU = tp_IoU / (tp_IoU + fn_IoU) if (tp_IoU + fn_IoU) > 0 else 0.0
    f1_score_IoU = 2 * (precision_IoU * recall_IoU) / (precision_IoU + recall_IoU + 1e-6)
    
    # Calculate Mean IoU
    if len(ious) > 0:
        mean_iou = np.mean(ious)
    else:
        mean_iou = 0.0
    
    # Store IoU metrics
    precision_IoU_list.append(precision_IoU)
    recall_IoU_list.append(recall_IoU)
    f1_score_IoU_list.append(f1_score_IoU)
    mean_iou_list.append(mean_iou)
    
    # Calculate Precision, Recall, and F1 Score for SIoU
    precision_SIoU = tp_SIoU / (tp_SIoU + fp_SIoU) if (tp_SIoU + fp_SIoU) > 0 else 0.0
    recall_SIoU = tp_SIoU / (tp_SIoU + fn_SIoU) if (tp_SIoU + fn_SIoU) > 0 else 0.0
    f1_score_SIoU = 2 * (precision_SIoU * recall_SIoU) / (precision_SIoU + recall_SIoU + 1e-6)
    
    # Calculate Mean SIoU
    if len(siou_list) > 0:
        mean_siou = np.mean(siou_list)
    else:
        mean_siou = 0.0
    
    # Store SIoU metrics
    precision_SIoU_list.append(precision_SIoU)
    recall_SIoU_list.append(recall_SIoU)
    f1_score_SIoU_list.append(f1_score_SIoU)
    mean_siou_list.append(mean_siou)

# Plotting the metrics
plt.figure(figsize=(12, 8))

# Plot Precision
plt.subplot(2, 2, 1)
plt.plot(thresholds, precision_IoU_list, label='Precision IoU', marker='o')
plt.plot(thresholds, precision_SIoU_list, label='Precision SIoU', marker='x')
plt.xlabel('IoU/SIoU Threshold')
plt.ylabel('Precision')
plt.title('Precision vs IoU/SIoU Threshold')
plt.legend()
plt.grid(True)

# Plot Recall
plt.subplot(2, 2, 2)
plt.plot(thresholds, recall_IoU_list, label='Recall IoU', marker='o')
plt.plot(thresholds, recall_SIoU_list, label='Recall SIoU', marker='x')
plt.xlabel('IoU/SIoU Threshold')
plt.ylabel('Recall')
plt.title('Recall vs IoU/SIoU Threshold')
plt.legend()
plt.grid(True)

# Plot F1 Score
plt.subplot(2, 2, 3)
plt.plot(thresholds, f1_score_IoU_list, label='F1 Score IoU', marker='o')
plt.plot(thresholds, f1_score_SIoU_list, label='F1 Score SIoU', marker='x')
plt.xlabel('IoU/SIoU Threshold')
plt.ylabel('F1 Score')
plt.title('F1 Score vs IoU/SIoU Threshold')
plt.legend()
plt.grid(True)

# Plot Mean IoU and Mean SIoU
plt.subplot(2, 2, 4)
plt.plot(thresholds, mean_iou_list, label='Mean IoU', marker='o')
plt.plot(thresholds, mean_siou_list, label='Mean SIoU', marker='x')
plt.xlabel('IoU/SIoU Threshold')
plt.ylabel('Mean IoU / SIoU')
plt.title('Mean IoU and SIoU vs IoU/SIoU Threshold')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
